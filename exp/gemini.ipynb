```python
from IPython.display import display
from IPython.display import Markdown

import pathlib
import textwrap
```

```python
import getpass
```

```python
apiKey= getpass.getpass("API KEY: ")
```

```python
def to_markdown(text):
    text=text.replace('•'," *")
    return Markdown(textwrap.indent(text,'> ',predicate=lambda _:True))
```

```python
import google.generativeai as genai

genai.configure(api_key=apiKey)
```

```python
model=genai.GenerativeModel('gemini-pro')
response=model.generate_content("What is the meaning of life?")
to_markdown(response.text)
```

```python
model=genai.GenerativeModel('gemini-pro')
chat=model.start_chat(history=[])
chat
```

```python
response=chat.send_message("In one sentence, explain how a computer works to a yound child")
to_markdown(response.text)
```

```python
response=chat.send_message("okay, how about a more detailed explanation to a high schooler?")
to_markdown(response.text)
```

```python
import google.ai.generativelanguage as glm
```

```python
model= genai.GenerativeModel('gemini-pro-vision')
response=model.generate_content(
    glm.Content(
    parts=[
        glm.Part(text="write a short, engaging blog post based on this picture going into every detail"),
        glm.Part(
            inline_data=glm.Blob(
                mime_type='image/jpeg',
                data=pathlib.Path('image2.jpg').read_bytes()
                )
            ),
        ],
    )
)

to_markdown(response.text[:100]+"...[TRIMMED]...")
```

```python
model=genai.GenerativeModel('gemini-pro')

messages=[
    {"role":"user",
     "parts":["Briefly explain how a computer work to a young child."]
    }
]
response= model.generate_content(messages)
to_markdown(response.text)
```

```python
messages.append({'role':'model',
                 'parts':[response.text]
                })
messages.append({
    'role':'user',
    'parts':["Okay, how about a more detailed explanation to a high school student?"]
})

response=model.generate_content(messages)

to_markdown(response.text)
```

```python
reponse=model.generate_content(
'tell me a story about a magic backpack',
    generation_config=genai.types.GenerationConfig(
    candidate_count=1,
    stop_sequences=['x'],
    max_output_tokens=20,
    temperature=1.0)
)
```
